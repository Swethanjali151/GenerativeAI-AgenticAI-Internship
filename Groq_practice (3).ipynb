```python
import os
from groq import Groq

os.environ["GROQ_API_KEY"]="gsk"

client = Groq()
completion = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
      {
        "role": "user",
        "content": "About jib openings in 2026"
      }
    ],
    temperature=1,
    max_completion_tokens=1024,
    top_p=1,
    stream=True,
    stop=None
)

for chunk in completion:
    print(chunk.choices[0].delta.content or "", end="")
```

```python

import os
from groq import Groq

client = Groq()
# Using a raw string to prevent unicode escape errors.
# Note: This is a local Windows path and will not be accessible directly in Colab.
# You need to upload your audio file to Colab or mount Google Drive and use the appropriate path.
filename = r"/content/Namo-Re.mp3"

with open(filename, "rb") as file:
    transcription = client.audio.transcriptions.create(
      file=(filename, file.read()),
      model="whisper-large-v3",
      temperature=0,
      response_format="verbose_json",
    )
    print(transcription.text)
```
