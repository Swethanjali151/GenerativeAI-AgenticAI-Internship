```python
df=pd.read_csv(r'C:\Users\sweth\Avscode\GenerativeAi_Internship\Python\titanic_ dataset_final.csv')
print(df.shape)
df.head()
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0
)
```

```python
agent = create_pandas_dataframe_agent(
    llm,
    df,
    verbose=True,
    allow_dangerous_code=True
)
```

```python
agent.invoke(
    "What are the top 5 records?"
)
```

```python
agent.run("How many people have more than 3 siblings")
```
